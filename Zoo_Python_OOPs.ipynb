For this project,we will create an ecosystem that represents a zoo. The zoo will have a
variety of animals and birds. These animals and birds can be organized as shown in the
following chart.
Each of these animal groups have some special characteristics that get passed on to the child.
The child along with the characteristics derived from the parent will also have their own
characteristics. This zoo can save a fixed number of animals and birds and is open to see the
animals in them.

Requirements:
1. Each animal has some common features like the number of hands and legs.
a. Felines and canines have 4 legs and no hands
2. Each bird has a number of legs and number of wings as a feature.
a. Flight birds have 2 legs and 2 wings.
Represent these features as attributes.
4. The characteristic for each of the animals are as follows
a. Felines belong to the cat family.
b. Canines belong to the dog family.
c. Tigers can roar and are lethal predators
d. Wild cats can climb trees
e. Wolves hunt in packs and have a leader
f. Flight birds fly and hunt for food
g. Eagles fly extremely high and can see their prey from high up in the sky.
5. Add the characteristics of animals and birds to their respective classes.
6. Create a zoo that can have 2 animals and 1 bird.
7. Zoo should be able to add only an animal or a bird if it is not full
8. Zoo should be able to provide a way to look at all the animals/birds it has.
9. Looking at animals/birds means you should be able to get all the features and
characteristics of them!

```python
class Animal:
    #Initialising the hands and legs of the animal
    def __init__(self, legs, hands=0):
        self.legs = legs
        self.hands = hands
   #Function to display the hands and legs of the animal
    def retrieve_characteristics(self):
        return f"There are {self.legs} legs and {self.hands} hands for this animal."


class Bird:
    #Initialising the hands and legs of the Bird
    def __init__(self, legs, wings):
        self.legs = legs
        self.wings = wings
    #Function to display the hands and legs of the Bird
    def retrieve_characteristics(self):
        return f"There are {self.legs} legs and {self.wings} wings for this bird."


class FlightBird(Bird):
   #Initialising the flight birds to have 2 legs and 2 wings
    def __init__(self):
        super().__init__(2, 2)
    #Displaying the features of bird along with special features of flightbird
    def retrieve_characteristics(self):
        return super().retrieve_characteristics() + " They can fly and hunt for food."


class Eagle(FlightBird):
   #Initialising the Eagles legs and wings
    def __init__(self):
        super().__init__()
  #Displaying the features of bird, flightbird along with special features of eagle
    def retrieve_characteristics(self):
        return super().retrieve_characteristics() + " They fly extremely high and can see their prey from high up in the sky."


class Feline(Animal):
    #Initialising the Felines to have 4 legs and no hands
    def __init__(self):
        super().__init__(4, 0)
    #Displaying the features of Animal along with special features of Feline
    def retrieve_characteristics(self):
        return super().retrieve_characteristics() + " Belongs to the cat family."


class Canine(Animal):
  #Initialising the Canines to have 4 legs and no hands
    def __init__(self):
        super().__init__(4, 0)
  #Displaying the features of Animal along with special features of Canine
    def retrieve_characteristics(self):
        return super().retrieve_characteristics() + " Belongs to the dog family."


class Tiger(Feline):
  #Initialising the Tiger's legs and hands
    def __init__(self):
        super().__init__()
   #Displaying the features of Animal, Feline along with special features of Tiger
    def retrieve_characteristics(self):
        return super().retrieve_characteristics() + " It can roar and are lethal predators."


class WildCat(Feline):
  #Initialising the Wildcat's legs and hands
    def __init__(self):
        super().__init__()
   #Displaying the features of Animal, Feline along with special features of Wildcat
    def retrieve_characteristics(self):
        return super().retrieve_characteristics() + " It can climb trees."


class Wolf(Canine):
  #Initialising the Wolf's legs and hands
    def __init__(self):
        super().__init__()
    #Displaying the features of Animal, Canine along with special features of Wolf
    def retrieve_characteristics(self):
        return super().retrieve_characteristics() + " They hunt in packs and have a leader."


class Zoo:
   #Initialising two empty lists of animals and bird that are to be added to the zoo
    def __init__(self):
        self.animals = []
        self.birds = []
   #Function to check if the zoo can accommodate more animals(max only 2) and if the added object is an Animal instance
    def add_animal(self, animal):
        if len(self.animals) < 2 and isinstance(animal, Animal):
            self.animals.append(animal)
            print(f"{type(animal).__name__} added to the zoo")
        else:
            print("Zoo is full for animals")

    def add_bird(self, bird):
      # Check if the zoo can accommodate more birds(max 1) and if the added object is a bird instance
        if len(self.birds) < 1 and isinstance(bird, Bird):
            self.birds.append(bird)
            print(f"{type(bird).__name__} added to the zoo")
        else:
            print("Zoo is full for birds")

    def show_animals(self):
      # Display characteristics of animals in the zoo
        for animal in self.animals:
            if isinstance(animal, Animal):
                print(animal.retrieve_characteristics())

    def show_birds(self):
      # Display characteristics of birds in the zoo
        for bird in self.birds:
            if isinstance(bird, Bird):
                print(bird.retrieve_characteristics())

# Function to get the animal to be added to the zoo as input from the user and creating instances appropriately of the animal
def create_animal():
    animal_type = input("Enter the type of animal (Tiger, WildCat, Wolf): ").strip().lower()
    if animal_type == "tiger":
        return Tiger()
    elif animal_type == "wildcat":
        return WildCat()
    elif animal_type == "wolf":
        return Wolf()
    else:
        print("Invalid animal type.")
        return None


zoo = Zoo()
eagle = Eagle()

#Bird is added to the zoo. Note: Input is not obtained from the user for bird as currently there is only one variety of bird that is acceptable(Eagle). A function create_bird may be created to extend to more bird varieties
zoo.add_bird(eagle)

#Trying to add two birds with the below function call to check if appropriate zoo full for birds message is displayed
zoo.add_bird(eagle)

# Loop to keep accepting the input from the user until conditions are satisfied and user enters done
while True:
    choice = input("Enter A to Add an animal and 'done' to finish): ").strip().lower()

    if choice == "a":
        new_animal = create_animal()
        if new_animal:
            zoo.add_animal(new_animal)
    elif choice == "done":
        break
    else:
        print("Invalid choice. Please enter 'A' or 'done'.")

print("\nAnimals in the zoo:")
zoo.show_animals()
print("\nBirds in the zoo:")
zoo.show_birds()
```
